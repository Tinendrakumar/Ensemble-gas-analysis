```python
import pandas
import numpy
import data_massage
import matplotlib.pyplot as plt
```

# Load and coerce data into generally useful forms
## Some models may further tailor the data to them

```python
metadata, full_data, induction_only = data_massage.load_data()
summarized_data = data_massage.summarize_data(metadata, induction_only)
```

```python
for label in ["banana", "wine", "background"]:
    labeled = summarized_data[summarized_data["class"] == label]
    plt.title("summaries for {}".format(label))
    
    #for feature in ["time_to_gas_max", "R_mean_mean", "Temp_mean", "Humidity_mean"]:
    for feature in ["R_mean_mean", "Temp_mean", "Humidity_mean"]:
        data = labeled[feature]
        plt.plot(data, label=feature)
    
    plt.legend()
    plt.show()
```

```python
for ind in metadata:
    induction = induction_only[induction_only["induction"] == ind]
    plt.title("{} {}".format(ind, metadata[ind]["class"]))
    
    for key in ["R_mean", "Temp.", "Humidity"]:
        plt.plot(induction[key], label=key)
    
    #plt.plot(metadata[ind]["time_to_gas_max"], label="time to gas max")
    
    plt.legend()
    plt.show()
```
