```python
import matplotlib.pyplot as plt
import numpy
import data_massage
from mpl_toolkits.mplot3d import Axes3D
np = numpy
```

```python
data = {
      "svm_summarized": [0.73, 0.57, 0.57, 0.65]
    , "7nn_summarized": [0.50, 0.53, 0.44, 0.50]
    , "7nn_full": [0.99, 0.96, 0.96, 0.97]
    , "7nn": [1.0, 0.99, 0.99, 0.99]
    , "garch_svm": [0.99, 0.97, 0.99, 0.99]
    , "garch_knn": [1.0, 0.73, 0.56, 0.76]
    , "arima_svm": [1.0, 0.96, 0.99, 0.99]
    , "arima_knn": [1.0, 0.55, 0.50, 0.66]
}
```

```python
ax = plt.axes()
plt.title("summarized data model performance")
plt.grid()
plt.xlabel("metric")
plt.ylabel("value")

for model in data:
    if "summarized" in model:
        plt.plot(data[model], label=model)


ax.set_yticks(np.arange(0.0, 1.0, 0.05))
ax.set_xticks(list(range(4)))
ax.set_xticklabels(["f1-background", "f1-banana", "f1-wine", "accuracy"])
plt.legend()
plt.show()
```

```python
ax = plt.axes()
plt.title("full data model performance")
plt.grid()
plt.xlabel("metric")
plt.ylabel("value")

for model in data:
    if "full" in model:
        plt.plot(data[model], label=model)


ax.set_yticks(np.arange(0.0, 1.1, 0.05))
ax.set_xticks(list(range(4)))
ax.set_xticklabels(["f1-background", "f1-banana", "f1-wine", "accuracy"])
plt.legend()
plt.show()
```

```python
ax = plt.axes()
plt.title("induction only data model performance")
plt.grid()
plt.xlabel("metric")
plt.ylabel("value")

for model in data:
    if all(map(lambda _: _ not in model, ["full", "summarized"])):
        plt.plot(data[model], label=model)

    
ax.set_yticks(np.arange(0.0, 1.0, 0.05))
ax.set_xticks(list(range(4)))
ax.set_xticklabels(["f1-background", "f1-banana", "f1-wine", "accuracy"])
plt.legend()
plt.show()
```
